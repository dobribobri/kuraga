```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)
```

```python
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
plt.figure()
plt.scatter(X[:, 1], X[:, 2], c=y)
```

```python
plt.figure()
plt.scatter(X[:, 2], X[:, 3], c=y)
```

```python
plt.figure()
plt.scatter(X[:, 3], X[:, 0], c=y)
```

```python
##################################################

import numpy as np

r = []
for x in X:
    r.append( np.linalg.norm(X - x, axis=1) )
r = 1. / np.asarray(r)
r = np.where(np.isinf(r), 0, r)
```

```python
r
```

```python
np.save('adjmat.npy', r)
```

```python
omega = []
for x in X:
    omega.append( np.linalg.norm(x) )
omega = np.asarray(omega)
omega.shape
```

```python
np.save('omega.npy', omega)
```

```python
##################################################

import numpy as np

r = []
for x in X:
    r.append( np.linalg.norm(X - x, axis=1) )
r = 1. / (np.asarray(r) ** 2)
r = np.where(np.isinf(r), 0, r)

np.save('adjmat2.npy', r)
```
